```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

creating patterns


```python
pattern1 = [{'LOWER': 'solarpower'}]

pattern2 = [{'LOWER': 'solar'}, {'LOWER': 'power'}]

pattern3 = [{'LOWER': 'solar'}, {'IS_PUNCT': True}, {'LOWER': 'power'}]

matcher.add('SolarPower', [pattern1, pattern2, pattern3])
```

```python
doc = nlp(u'The Solar Power industry continues to grow as demand \
for solarpower increases. Solar-power cars are gaining popularity.')
```

```python
found_matches = matcher(doc)
print(found_matches)
```

```python
for match_id, start, end in found_matches:
  string_id = nlp.vocab.strings[match_id]
  span = doc[start:end]
  print(match_id, string_id, start, end, span.text)
```

```python
# redefine the pattern
pattern1 = [{'LOWER': 'solarpower'}]
pattern2 = [{'LOWER': 'solar'}, {'IS_PUNCT': True, 'OP':'*'}, {'LOWER': 'power'}]
```

```python
#add new set
matcher.add('SolarPower',[pattern1,pattern2])
```

```python
found_matches = matcher(doc)
print(found_matches)
```

```python
pattern1 = [{'LOWER': 'solarpower'}]
pattern2 = [{'LOWER': 'solar'}, {'IS_PUNCT': True, 'OP':'*'}, {'LEMMA': 'power'}]

#REMOVE OLD PATTERN
matcher.remove('SolarPower')

#ADD NEW PATTERN
matcher.add('SolarPower', [pattern1, pattern2])
```

```python
doc2 = nlp(u'Solar-powered energy runs solar-powered cars.')
```

```python
found_matches = matcher(doc2)
print(found_matches)
```

PhraseMatcher


```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)
```

```python
with open('/content/reaganomics.txt','r', encoding='latin-1') as f:
  doc3 = nlp(f.read())
```

```python
#First, create a list of match phrase
phrase_list = ['voodoo economics', 'supply-side economics', 'trickle-down economics', 'free-market economics']
```

```python
# next, convert each phrase to a doc object
phrase_patterns = [nlp(text) for text in phrase_list]
```

```python
#pass each doc object into matcher
matcher.add('VoodooEconomics', None, *phrase_patterns)
```

```python
# build a list of a matches:
matches = matcher(doc3)
```
