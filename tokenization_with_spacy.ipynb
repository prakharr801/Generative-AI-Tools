```python
# import library spacy
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
mystring = '"we\'re moving to l.a.!"'
print(mystring)
```

```python
# create a doc object and explore tokens.
doc = nlp(mystring)

for token in doc:
    print(token.text,end='|')
```

```python
#prefixes , suffixes , and infxies
doc2=nlp(u"we're here to help! send snail-mail ,email support@oursite.com or visit us at http:\\www.oursite.com!")
for t in doc2:
    print(t)
```

```python
doc3 =nlp(u'a 5km nyc cab ride costs $10.30')
for t in doc3:
    print(t)
```

```python
#exceptions
doc4=nlp(u"let's visit st.louis in the u.s next year.")
for t in doc4:
    print(t)
```

```python
# token can be retrived index position and slicing
doc5 = nlp(u'it is better to give than to receive.')
#retrive the third token
doc5[2]
#retrive the token from middle
doc5[2:5]
# retrive the token from lastfour.
doc5[-2]
doc5[-4:]
```

```python
# token cannot be reassigned
doc6 = nlp(u'my dinner was horrible.')
doc7 = nlp(u'your dinner was delicious.')

# You cannot reassign tokens like this:
# doc6[3] = doc7[3]

# Instead, you can access tokens by their index:
print(f"Token at index 3 in doc6: {doc6[3]}")
print(f"Token at index 3 in doc7: {doc7[3]}")
```

```python
# named entities
doc8 = nlp(u'Apple to bulid a hong kong factory for $6 million')
for token in doc8:
   print(token.text,end='|')
print('\n----')

for ent in doc8.ents:
    print(ent.text+'-'+ent.label_+'-'+str(spacy.explain(ent.label_)))
```
