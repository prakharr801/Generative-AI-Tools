6. Write a matcher called 'Swimming' that finds both occurrences of the phrase "swimming vigorously" in the text.

```python
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("en_core_web_sm")

text = open("/content/owlcreek.txt").read()
doc = nlp(text)


matcher = Matcher(nlp.vocab)
pattern = [{"LOWER": "swimming"}, {"LOWER": "vigorously"}]
matcher.add("Swimming", [pattern])

matches = matcher(doc)

if matches:
    for match_id, start, end in matches:
        span = doc[start:end]
        print(f"Match found: '{span.text}' at position {start}-{end}")
else:
    print("No matches found ")
```

7. Print the text surrounding each found match.

```python
# Apply the matcher to the doc
matches = matcher(doc)

# Print sentence context around each match
for match_id, start, end in matches:
    span = doc[start:end]
    sentence = span.sent.text
    print(f"Match: '{span.text}'")
    print(f"Sentence: {sentence}\n")
```
