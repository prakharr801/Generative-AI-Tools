```python
f=open("BMW-Group-Report-2024-en (1).pdf","rb")
pdf_reader=PyPDF2.PdfReader(f)
```

```python
second_page=pdf_reader.pages[1]
```

```python
pdf_writer=PyPDF2.PdfWriter(f)
```

```python
pdf_writer.add_page(second_page)
```

```python
pdf_output=open('new_doc.pdf','wb')
```

```python
pdf_writer.write(pdf_output)
```

```python
pdf_output.close()
f.close()
```
